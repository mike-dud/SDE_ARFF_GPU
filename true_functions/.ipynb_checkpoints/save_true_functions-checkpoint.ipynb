```python
import os
import dill
import jax.numpy as jnp

# true functions parameters
n_dimension = 1

def true_drift(x):
    def f_1(x):
        A = -1.5
        B = jnp.array([[4], [0]])
        C = jnp.array([[0], [0]])
        D = jnp.array([[-2], [0]])
        f = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return f
    
    def f_2(x):
        A = -1.5
        B = jnp.array([[0], [4]])
        C = jnp.array([[0], [0]])
        D = jnp.array([[0], [-2]])
        f = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return f

    drift = jnp.transpose(jnp.array([f_1(x), f_2(x)]))
    return drift


def true_diffusion(x):
    def sigma_11(x):
        A = 0.5  
        B = jnp.array([[0.05], [0]])
        C = jnp.array([[0], [0]])
        D = jnp.array([[0], [0]])
        sigma = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return sigma
    
    def sigma_12(x):
        A = 0
        B = jnp.array([[0], [0]])
        C = jnp.array([[0], [0]])
        D = jnp.array([[0], [0]])
        sigma = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return sigma
    
    def sigma_22(x):
        A = 0.5
        B = jnp.array([[0], [0.05]])
        C = jnp.array([[0], [0]])
        D = jnp.array([[0], [0]])
        sigma = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return sigma

    diffusion = jnp.abs(jnp.transpose(jnp.array([[sigma_11(x), sigma_12(x)], [sigma_12(x), sigma_22(x)]])))
    return diffusion


functions = {
    "drift": true_drift,
    "diffusion": true_diffusion,
}

with open("ex3-2D_cubic.pkl", "wb") as f:
    dill.dump(functions, f)
```
