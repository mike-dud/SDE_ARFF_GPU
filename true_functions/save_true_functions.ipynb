```python
import os
import dill
import jax.numpy as jnp

def true_drift(x):
    def f_1(x):
        A = 0
        B = jnp.array([[-1], [0], [0]])
        C = jnp.array([[0], [0], [0]])
        D = jnp.array([[0], [0], [0]])
        f = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return f

    def f_2(x):
        A = 0
        B = jnp.array([[0], [-1], [0]])
        C = jnp.array([[0], [0], [0]])
        D = jnp.array([[0], [0], [0]])
        f = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return f

    def f_3(x):
        A = 0
        B = jnp.array([[0], [0], [-1]])
        C = jnp.array([[0], [0], [0]])
        D = jnp.array([[0], [0], [0]])
        f = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return f

    drift = jnp.transpose(jnp.array([f_1(x), f_2(x), f_3(x)]))
    return drift


def true_diffusion(x):
    def sigma_11(x):
        A = 0.02045836
        B = jnp.array([[0], [0], [0]])
        C = jnp.array([[0], [0], [0]])
        D = jnp.array([[0], [0], [0]])
        sigma = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return sigma

    def sigma_12(x):
        A = -0.03502162
        B = jnp.array([[0], [0], [0]])
        C = jnp.array([[0], [0], [0]])
        D = jnp.array([[0], [0], [0]])
        sigma = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return sigma

    def sigma_13(x):
        A = -0.02678421
        B = jnp.array([[0], [0], [0]])
        C = jnp.array([[0], [0], [0]])
        D = jnp.array([[0], [0], [0]])
        sigma = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return sigma

    def sigma_22(x):
        A = 0.06355898
        B = jnp.array([[0], [0], [0]])
        C = jnp.array([[0], [0], [0]])
        D = jnp.array([[0], [0], [0]])
        sigma = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return sigma

    def sigma_23(x):
        A = 0.02981971
        B = jnp.array([[0], [0], [0]])
        C = jnp.array([[0], [0], [0]])
        D = jnp.array([[0], [0], [0]])
        sigma = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return sigma

    def sigma_33(x):
        A = 0.12453806
        B = jnp.array([[0], [0], [0]])
        C = jnp.array([[0], [0], [0]])
        D = jnp.array([[0], [0], [0]])
        sigma = jnp.squeeze(A + jnp.dot(x, B) + jnp.dot(jnp.square(x), C) + jnp.dot(jnp.power(x, 3), D))
        return sigma

    diffusion = jnp.abs(jnp.transpose(jnp.array(
        [[sigma_11(x), sigma_12(x), sigma_13(x)], 
         [sigma_12(x), sigma_22(x), sigma_23(x)],
         [sigma_13(x), sigma_23(x), sigma_33(x)]])))
    return diffusion


functions = {
    "drift": true_drift,
    "diffusion": true_diffusion,
}

with open("ex4-3D_spd.pkl", "wb") as f:
    dill.dump(functions, f)
```
